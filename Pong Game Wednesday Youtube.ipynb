```python
import turtle #basic grafic module,get starting with gamer, for beginner its simple 

wn = turtle.Screen()
wn.title("Pong by AISciences")
wn.bgcolor("black") 
wn.setup(width=800, height=600)
wn.tracer(0) #stop the window for updating we can manualing updating for speedup over game 


# Paddle A
paddle_a = turtle.Turtle() 
paddle_a.speed(0) #not a speed of padel that move on screen,its a speed of animation
paddle_a.shape("square") #buildin shape ,have multiple shape but we are using square
paddle_a.color("white") # paddle_a.shapesize(stretch_wid=5,stretch_len=1)
paddle_a.shapesize(stretch_wid=5,stretch_len=1) # change the size of paddle
paddle_a.penup() #by defalut turle draw a line but in game we cannot draw a line so penup
paddle_a.goto(-350, 0) # 0 is center of screen -350 X-axis


# Paddle B
paddle_b = turtle.Turtle()
paddle_b.speed(0)
paddle_b.shape("square")
paddle_b.color("white")
paddle_b.shapesize(stretch_wid=5,stretch_len=1)
paddle_b.penup()
paddle_b.goto(350, 0)


# Ball
ball = turtle.Turtle()
ball.speed(0)
ball.shape("circle") # now we can use circle shape
ball.color("white")
ball.penup()
ball.goto(0, 0) #center of screen
ball.dx = 0.5     #delta change or x speed or ball x-axis speed,every time ball move by 2 pixels
ball.dy = 0.5     #delta change or y speed or ball y-axis speed,every time ball move by 2 pixels

# Pen
pen = turtle.Turtle()
pen.speed(0)
pen.shape("square")
pen.color("white")
pen.penup()
pen.hideturtle() # we cannot show a object on screen
pen.goto(0, 260)
pen.write("Player A: 0  Player B: 0", align="center", font=("Courier", 24, "normal"))

# Score
score_a = 0
score_b = 0


# Functions
def paddle_a_up():
    y = paddle_a.ycor() # ycor() return the the y-axis coordinate
    y += 20             # add 20 pixcel to the y coordinate to move up
    paddle_a.sety(y)    # set y axis coordinate

def paddle_a_down():
    y = paddle_a.ycor()
    y -= 20             # subtract 20 pixcel to the y coordinate to move down      
    paddle_a.sety(y)

def paddle_b_up():
    y = paddle_b.ycor()
    y += 20              # add 20 pixcel to the y coordinate to move up
    paddle_b.sety(y)

def paddle_b_down():
    y = paddle_b.ycor()
    y -= 20               # subtract 20 pixcel to the y coordinate to move down 
    paddle_b.sety(y)


# Keyboard bindings
wn.listen()                           #listen the keyboard input
wn.onkeypress(paddle_a_up, "w")        #press w to move paddle a to up
wn.onkeypress(paddle_a_down, "s")
wn.onkeypress(paddle_b_up, "Up")
wn.onkeypress(paddle_b_down, "Down")




# Main Game Loop
while True:
    wn.update() #every loop iteration screen updated
    
     # Move the ball
    ball.setx(ball.xcor() + ball.dx)  #(ball.xcor() ball current x coordinate) ball start(0,0) ball move by 2 pixels
    ball.sety(ball.ycor() + ball.dy)

    # Border checking

    # Top and bottom
    if ball.ycor() > 290: # if ball current coordinate is greater than 290 
        ball.sety(290)    # again set ball location at 290 coordinate
        ball.dy *= -1     #reverse the direction of ball
    
    elif ball.ycor() < -290:
        ball.sety(-290)
        ball.dy *= -1

    # Left and right
    if ball.xcor() > 350:
        score_a += 1
        pen.clear()
        pen.write("Player A: {}  Player B: {}".format(score_a, score_b), align="center", font=("Courier", 24, "normal"))
        ball.goto(0, 0)
        ball.dx *= -1

    elif ball.xcor() < -350:
        score_b += 1
        pen.clear()
        pen.write("Player A: {}  Player B: {}".format(score_a, score_b), align="center", font=("Courier", 24, "normal"))
        ball.goto(0, 0)
        ball.dx *= -1


    # Paddle and ball collisions
    
    if ball.xcor() < -340 and ball.ycor() < paddle_a.ycor() + 50 and ball.ycor() > paddle_a.ycor() - 50:
        ball.dx *= -1 
    
    elif ball.xcor() > 340 and ball.ycor() < paddle_b.ycor() + 50 and ball.ycor() > paddle_b.ycor() - 50:
        ball.dx *= -1
    
```
